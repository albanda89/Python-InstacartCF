```python
# 1st method -aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# creating a loyalty flag for existing customers using the transform() and loc() functions.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False).sum()
```

```python
#check the basic statistics of the product prices for each loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# The prices of products purchased by loyal customers are cheaper than that of both categories, new customers and Regular customers purchase comparatively higher-cost products, and among the most expensive are purchased by new customers.

```python
# mean_prices - customers spending habits identification

ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] <10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'Higher spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False).sum()
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'prices']].head(30)
```

```python
# median_days - customers buying frequency habits identification

ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] >= 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] >= 10) & (ords_prods_merge['median_days'] < 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
